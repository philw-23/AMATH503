```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf
plt.rcParams['font.size'] = '14'
```

```python
# Function for problem 1
# NOTE: Left in this form for code as was running into overflow error when using simplified version
def problem1Func(x, t):
    integral_1 = 2*np.sqrt(np.pi * t)*(x + 1)*(erf((x + 1)/(4*np.sqrt(t))) - erf(x/(4*np.sqrt(t)))) + \
                    8*t*(np.exp(-(x + 1)**2 / (16*t)) - np.exp(-x**2 / (16 * t)))
    integral_2 = 2*np.sqrt(np.pi * t)*(x - 1)*(erf((x - 1)/(4*np.sqrt(t))) - erf(x/(4*np.sqrt(t)))) + \
                    8*t*(np.exp(-(x - 1)**2 / (16*t)) - np.exp(-x**2 / (16 * t)))
    full_sol = integral_1 + integral_2
    
    return (1 / (4*np.sqrt(np.pi * t))) * full_sol

    
```

```python
# Function for problem 2
def problem2Func(x, t):
    sinh = 2*np.sinh(x)
    erf_one = np.exp(x) * erf((2*t + x)/(2*np.sqrt(t)))
    erf_two = np.exp(-x) * erf((2*t - x)/(2*np.sqrt(t)))
    full_sol = sinh - erf_one + erf_two

    return np.exp(t) * 0.5 * full_sol
```

```python
# Problem 1
x = np.linspace(-4, 4, 1000)
t_vals = [0.00001, 0.01, 0.1, 0.5]
fig = plt.figure(figsize=(15, 10))

for t in t_vals:
    sol = problem1Func(x, t)
    plt.plot(x, sol, label='$t$=' + str(t))

plt.grid('on')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$u(t, x)$')
plt.savefig('./Problem_1.png', bbox_inches='tight')
plt.show()
```

```python
# Problem 2
x = np.linspace(-2*np.pi, 0, 1000)
t_vals = [0, 0.25, 0.75, 1]
fig = plt.figure(figsize=(15, 10))

for t in t_vals:
    sol = problem2Func(x, t)
    plt.plot(x, sol, label='$t$=' + str(t))

plt.grid('on')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$u(t, x)$')
plt.savefig('./Problem_2.png', bbox_inches='tight')
plt.show()
```
