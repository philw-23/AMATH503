```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['font.size'] = '14'
```

```python
def problemOnePlot(x, t):
    f_x = 0.5*(np.exp(-(x + 3*t)**2) + np.exp(-(x - 3*t)**2)) # Sol from f(x) non-zero
    g_x = (1/6)*(np.arctan(x + 3*t) - np.arctan(x - 3*t)) # Sol from g(x) non-zero
    h_x = np.zeros(x.shape)
    h_x = np.where((x >= 0) & (x < 3*t), t/3 - x/9, np.where((x > -3*t) & (x < 0), t/3 + x/9, 0)) # Sol from h(x) non-zero
    return f_x + g_x + h_x
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
t_vals = [0, 0.1, 0.5, 1, 2]
fig = plt.figure(figsize=(15,10))

for t in t_vals:
    sol = problemOnePlot(x, t)
    plt.plot(x, sol, label='$t$=' + str(t))
    
plt.grid('on')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$u(t,x)$')
plt.savefig('./Problem_1.png', bbox_inches='tight')
plt.show()
```
