```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
```

# Problem 1

```python
x = np.linspace(0, np.pi, 1000)
u_tx = lambda n, t, x: 2*((-1)**n + 1)/(np.pi*(1 - n**2))*np.cos(2*n*t)*np.cos(n*x)
t_vals = [0, 0.25, 0.5, 0.75]
fig, ax = plt.subplots(1, 4, figsize=(22, 5))
for idx, t in enumerate(t_vals):
    sol = (2 / np.pi) * np.ones(x.shape)
    for n in range(2, 100 + 1):
        sol += u_tx(n, t, x)
    ax[idx].grid('on')
    ax[idx].plot(x, sol)
    ax[idx].set_xlabel('$x$')
    ax[idx].set_ylim([-2, 2])
    if idx == 0:
        ax[idx].set_ylabel('$u(t,x)$')
    ax[idx].set_title('t=' + str(t))

plt.savefig('./Problem_1.png', bbox_inches='tight')
```

# Problem 2

```python
x = np.linspace(0, 2, 1000)
t_vals = [0, 0.25, 0.5, 1]
u_inf = lambda n, x, t: 2*(1 - np.cos(np.pi*n))/(np.pi * n) * \
                        np.exp(-(np.pi*n/2)**2 * t) * np.sin(n*np.pi*x/2)

fig, ax = plt.subplots(1, 4, figsize=(22, 5))
for idx, t in enumerate(t_vals):
    base_sol = 1 + x
    for n_val in range(1, 100 + 1):
        next_term = u_inf(n_val, x, t)
        base_sol += next_term
        
    ax[idx].grid('on')
    ax[idx].plot(x, base_sol)
    ax[idx].set_xlabel('$x$')
    if idx == 0:
        ax[idx].set_ylabel('$u(t,x)$')
    ax[idx].set_title('t=' + str(t))
    
plt.savefig('./Problem_2.png', bbox_inches='tight')
```

# Problem 3

```python
x = np.linspace(-np.pi, np.pi, 1000)
f_x = lambda n, x: 1j*np.exp(1j*n*x) / (2*np.pi*n) * ((-1)**n - 1)
N_vals = [5, 10, 100, 1000]

fig, ax = plt.subplots(1, 4, figsize=(22, 5))
for idx, N in enumerate(N_vals):
    sol = np.zeros(x.shape, dtype='complex')
    for n in range(-N, N + 1):
        if n == 0:
            sol += 1/2 * np.ones(x.shape)
            continue
        else:
            sol += f_x(n, x)
    
    ax[idx].grid('on')
    ax[idx].plot(x, sol)
    ax[idx].set_xlabel('$x$')
    if idx == 0:
        ax[idx].set_ylabel('$f(x)$')
    ax[idx].set_title('N =' + str(N))

plt.savefig('./Problem_3_Base.png', bbox_inches='tight')
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
f_x = lambda n, x: ((-1)**n - 1) / (2*n**2 * np.pi) * np.exp(1j*n*x)
N_vals = [5, 10, 100, 1000]

fig, ax = plt.subplots(1, 4, figsize=(22, 5))
for idx, N in enumerate(N_vals):
    sol = np.zeros(x.shape, dtype='complex')
    for n in range(-N, N + 1):
        if n == 0:
            sol += 1/2 * x
            continue
        else:
            sol += f_x(n, x)
    
    ax[idx].grid('on')
    ax[idx].plot(x, sol)
    ax[idx].set_xlabel('$x$')
    if idx == 0:
        ax[idx].set_ylabel('$f(x)$')
    ax[idx].set_title('N =' + str(N))
plt.savefig('./Problem_3_Int.png', bbox_inches='tight')
```
