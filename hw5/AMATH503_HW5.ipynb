```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 16})
```

```python
# Generate data and plot
c = 1
t_vals = [0, 1, 2, 3, 5]
x = np.linspace(-np.pi, np.pi, 1000)
f = lambda c, t, x: (1/c) * (1 - np.exp(-c*t)*np.cosh(x))


fig = plt.figure(figsize=(15, 12))
for t in t_vals:
    plt.plot(x, f(c, t, x), label='t=' + str(t))
plt.grid()
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$u(t, x)$')
plt.savefig('./Problem_1_Plot.png', bbox_inches='tight')
```
